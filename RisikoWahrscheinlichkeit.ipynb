```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def würfeln(n,vert,ang):
    würfelv=np.zeros((n,3))
    würfela=np.zeros((n,3))
    
    gew_matrix2=np.zeros(n,)
    gew_matrix3=np.zeros(n,)
    gew_matrix4=np.zeros(n,)
    probabilityb_func=[]
    probabilityva_func=[]
    probabilityaa_func=[]
    sum_of_probabilities=0
    for i in range(n):
        
       
        for f in range(vert):
            würfelv[i,f]=np.random.randint(1,7)
        for k in range(ang):
            würfela[i,k]=np.random.randint(1,7)
    würfelv=np.sort(würfelv)
    würfela=np.sort(würfela)
    erg_matrix=würfelv-würfela
    
    for m in range(n):
        if vert==2 and ang>=2:
            if erg_matrix[m,1]>=0 and erg_matrix[m,2]>=0:
                gew_matrix3[m]=1
            elif erg_matrix[m,1]<0 and erg_matrix[m,2]<0:
                gew_matrix4[m]=1
            
            elif erg_matrix[m,2]>=0 and erg_matrix[m,1]<0:
                gew_matrix2[m]=1
                
            elif erg_matrix[m,2]<0 and erg_matrix[m,1]>=0:
                gew_matrix2[m]=1
                
        else:
            
            if erg_matrix[m,2]>=0:
                gew_matrix3[m]=1
            elif erg_matrix[m,2]<0:
                gew_matrix4[m]=1
        
        probabilityb_func.append((np.sum(gew_matrix2))/(m+1))
        #probabilitya_func.append((np.sum(gew_matrix1))/(m+1))
        probabilityva_func.append((np.sum(gew_matrix3))/(m+1))
        probabilityaa_func.append((np.sum(gew_matrix4))/(m+1))
        
        
    #print(np.sum(gew_matrix1))
    #print(np.sum(gew_matrix2))
    #print(np.sum(gew_matrix3))
    #print(np.sum(gew_matrix4))
    #print(np.sum(np.stack((gew_matrix2,gew_matrix3,gew_matrix4))))
    #plt.figure()
    #plt.subplot(211)
    #plt.scatter(np.arange(len(step_func)),step_func)
    
    
    
    plt.plot(np.arange(len(probabilityb_func)),probabilityb_func,label="Unentschieden")
    #plt.plot(np.arange(len(probabilitya_func)),probabilitya_func,label="Angreifer")
    plt.plot(np.arange(len(probabilityva_func)),probabilityva_func,label="Verteidiger gewinnt alles")
    plt.plot(np.arange(len(probabilityaa_func)),probabilityaa_func,label="Angreifer gewinnt alles")
    
    plt.ylabel("Wahrscheinlichkeit")
    plt.xlabel("Würfelversuche")
    plt.legend()
    return print("Die durchschnittliche Anzahl der gewonnenen Einheiten bei",
                 vert," Verteidigern und ",ang," Angreifern ist: Verteidiger: ",
                 float(probabilityb_func[n-1]+vert*probabilityva_func[n-1])," Angreifer: ",
                float(probabilityb_func[n-1]+vert*probabilityaa_func[n-1])),print(erg_matrix.shape),print(
        "Die Wahrscheinlichkeit, dass der Angreifer alles gewinnt: ",probabilityaa_func[n-1]),print(
        "Die Wahrscheinlichkeit, dass der Verteidiger alles gewinnt: ",probabilityva_func[n-1]),print(
    "Die Wahrscheinlichkeit für ein Unentschieden: ",probabilityb_func[n-1]),sum_of_probabilities==(probabilityaa_func[n-1]+probabilityva_func[n-1]+probabilityb_func[n-1])
```
