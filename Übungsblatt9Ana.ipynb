```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fourier(n,y):
    x=np.arange(-y,y)
        
    for i in np.arange(1,n):
        F_old=0
        F=np.sin(i*x)/i
        F_new=F_old+F
        
    return F_new
```

```python
plt.plot(fourier(1000,100))
    
```

```python
def gaussian(x,a):
    
    return np.exp((-x**2)/(2*a**2))
def gaussianderiv1(x,a):
    
    return -2/a**2*np.exp((-x**2)/(2*a**2))
def gaussianderiv2(x,a):
    
    return -4/a**4*np.exp((-x**2)/(2*a**2))
```

```python
zeh=[]
neh=[]
keh=[]
ding=np.arange(-50,50)
for i in ding:
    zeh.append(gaussian(i,4))
    neh.append(gaussianderiv1(i,4))
    keh.append(gaussianderiv2(i,4))
```

```python
plt.plot(ding,zeh)
plt.plot(ding,neh)
plt.plot(ding,keh)
```

```python
def komich(x):
    return (np.exp(x)-np.sin(x)-1)/(x**2)
```

```python
def ka(x):
    if x==0:
        return
    else:
        
        return 2*np.sin(1/x)+np.cos(1/x)
```

```python
sacktaste=[]
for i in ding:
    sacktaste.append(ka(i))
    
```
